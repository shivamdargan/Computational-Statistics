```python
import numpy as np
import pandas as pd

data = pd.read_csv(r'./kc_house_data.csv')
# data
df = pd.DataFrame(data, columns=['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_above'])
df
```

```python
def R_square(x1,x2):
    X1_X2_sum = 0
    X1_square_sum = 0
    X2_square_sum = 0
    sum_x1 = np.sum(x1)
    sum_x2 = np.sum(x2)
    n = len(df.iloc[:,4].values)
    for i in range(n):
        X2_square_sum += x2[i]**2
        X1_square_sum += x1[i]**2
        X1_X2_sum += x1[i]*x2[i]
    
    num = (n*X1_X2_sum) - (sum_x1*sum_x2)
    den = (((n*X1_square_sum) - (sum_x1**2))**0.5)*(((n*X2_square_sum) - (sum_x2)**2)**0.5)
    return (num/den)**2
```

```python
def VIF(r_square):
    vif = 1/(1-r_square)
    return vif
```

# Calculating Factors Like VIF, Correlation

```python
for i in range(1,6):
    print("\033[1m","Correlation","\033[0m")
    for j in range(i,6):
        ColRelation = "Between Column "+str(i-1)+" & Column"+str(j)+": "
        print(ColRelation,R_square(df.iloc[:,i-1].values,df.iloc[:,j].values).round(4))
    print('\n')
```

```python
colsDropped=[]
for i in range(1,6):
    print("\033[1m","Variance Inflation Factor","\033[0m")
    for j in range(i,6):
        ColRelation = "Between Column "+str(i-1)+" & Column"+str(j)+": "
        vifCalc=VIF(R_square(df.iloc[:,i-1].values,df.iloc[:,j].values).round(4))
        print(ColRelation,vifCalc)
        if(vifCalc>4):
            colsDropped.append(i)
    print('\n')
```

## Dropping the Columns Which Have High VIF Between Them

```python
for i in range(len(colsDropped)):
    df.drop(df.columns[colsDropped[i]-1],axis=1, inplace=True)      
df
```
