
# Calculated Using Formula's - User Defined Functions


```python
def sumOfData():
    sum=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            sum=sum+x[i][j]
    print('Sum Of All Fields:', sum)
```

```python
def meanOfData():
    row_mean,column_mean=[],[]
    for i in range(x.shape[1]):
        sum_column=0
        for j in range(x.shape[0]):
            sum_column=sum_column+x[j][i]
        column_mean.append((sum_column/x.shape[0]))
    print('mean of All Columns Are: ', column_mean)
    for i in range(x.shape[0]):
        sum_rows=0
        for j in range(x.shape[1]):
            sum_rows=sum_rows+x[i][j]
        row_mean.append((sum_rows/x.shape[1]))
    print('mean of All Rows Are: ', row_mean)
    print("**********************")
    return(column_mean,row_mean)
```

```python
def varianceStdDevOfData(column_mean,row_mean):
    column_var=[]
    for i in range(x.shape[1]):
        var_column=0
        for j in range(x.shape[0]):
            var_column=var_column+(x[j][i]-column_mean[i])**2
        column_var.append((var_column/x.shape[0]))
    print('Variance of All Columns Are: ', column_var)
    print('Standard Deviation of All Columns Are: ', np.sqrt(column_var))
    row_var=[]
    for i in range(x.shape[0]):
        var_row=0
        for j in range(x.shape[1]):
            var_row=var_row+(x[i][j]-row_mean[i])**2
        row_var.append((var_row/x.shape[1]))
    print('Variance of All Rows Are: ', row_var)
    print('Standard Deviation of All Rows Are: ', np.sqrt(row_var))
    print("**********************")
    return column_var
```

```python
def medianOfData():
    row_median,column_median=[],[]
    for i in range(x.shape[1]):
        col_elem,row_elem=[],[]
        for j in range(x.shape[0]):
            col_elem.append(x[j][i])
        col_elem.sort()
        if(len(col_elem)%2==0):
            column_median.append((col_elem[math.floor((len(col_elem)-1)/2)]+col_elem[math.floor((len(col_elem)+1)/2)])/2)
        else:
             column_median.append(col_elem[math.floor(len(col_elem)/2)])
    print('median of All Columns Are: ', column_median)
    for i in range(x.shape[0]):
        col_elem,row_elem=[],[]
        for j in range(x.shape[1]):
            row_elem.append(x[i][j])
        row_elem.sort()
        if((len(row_elem)%2)==0):
            row_median.append((row_elem[math.floor((len(row_elem)-1)/2)]+row_elem[math.floor((len(row_elem)+1)/2)])/2)
        else:
            row_median.append(row_elem[math.floor(len(row_elem)/2)])
    print('median of All Rows Are: ', row_median)
    print("**********************")
```

```python
def covarianceOfData(column_mean,row_mean):
    for i in range(x.shape[1]-1):
        cov=0
        for j in range(x.shape[0]):
            cov=cov+((x[j][i]-column_mean[i])*(x[j][i+1]-column_mean[i+1]))
        covar.append(cov/x.shape[0]-1)
    print('covariance by formula is ', covar)
    print("**********************")
    return covar
```

```python
def correlationOfData(column_var,cov):
    corr=[]
    for i in range(x.shape[1]-1):
        corr.append(cov[i]/(math.sqrt(column_var[i]) * math.sqrt(column_var[i+1])))
    print('Correlation Is', corr)
    print("**********************")
```

```python
while(True):
    print("Select One From The Choice: ")
    print("1) Sum Of Data")
    print("2) Mean Of Data")
    print("3) Variance Of Data")
    print("4) Co-Variance Of Data")
    print("5) Correlation Of Data")
    print("6) Median Of Data")
    print("7) Exit")
    option = int(input('Choose an option: '))
    if option == 1:
        sumOfData()
    elif option == 2:
        result=meanOfData()
    elif option == 3:
        res=varianceStdDevOfData(result[0],result[1])
    elif option == 4:
        covar=covarianceOfData(result[0],result[1])
    elif option == 5:
        correlationOfData(res,covar)
    elif option == 6:
        medianOfData()
    elif option == 7:
        break
```
