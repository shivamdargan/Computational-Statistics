```python
df=pd.read_csv(r'./kc_house_data.csv')
df
```

```python
df=df.iloc[:,[5,2,6]]
df
df.dropna()
```

# Normalization Of Dataset

```python
max_val = []
min_val = []
for c in df.columns:
    max_val.append(df[c].max())
    min_val.append(df[c].min())
temp = 0
for c in df.columns:
    while(temp < len(df.columns)):
        df[c] = (df[c] - min_val[temp])/(max_val[temp]-min_val[temp])
        temp = temp+1
        break
        
df
```

# Multiple linear Regression model
### price=$B_0$+$B_1$(sqft_living)+$B_2$(sqft_lot)

```python
x1=df.iloc[:,0]   #sqft_living
x2=df.iloc[:,1]   #sqft_lot
y=df.iloc[:,2] #price
```

## Computing All Of The Parameters

```python
x12=0
x22=0
x1_sum=0   #summation of sqft_living column 
x2_sum=0   #summation of sqft_lot column 
y_sum=0    #summation of y column 
x1_y=0     #summation of multiplication of sqft_living and y column
x2_y=0     #summation of multiplication of sqft_lot and y column
x1_x2=0    #summation of multiplication of sqft_living and sqft_lot column
for i in range(0,df.shape[0]):
    x12+=x1[i]**2
    x22+=x2[i]**2
    x1_sum+=x1[i]
    x2_sum+=x2[i]
    y_sum+=y[i]
    x1_y+=x1[i]*y[i]
    x2_y+=x2[i]*y[i]
    x1_x2+=x1[i]*x2[i]
```

```python
x1_mean=0    #mean of sqft_living column
x2_mean=0    #mean of sqft_lot column
y_mean=0     #mean of y column
x1_mean=x1_sum/df.shape[0]
x2_mean=x2_sum/df.shape[0]
y_mean=y_sum/df.shape[0]
```

# Initializing All Parameters

```python
sum_x12=0
sum_x22=0
sum_x1y=0
sum_x2y=0
sum_x1x2=0
```

## Applying Computational Formulas For MLR

```python
sum_x12=x12-(x1_sum**2)/df.shape[0]
sum_x22=x22-(x2_sum**2)/df.shape[0]
sum_x1y=x1_y-((x1_sum)*(y_sum))/df.shape[0]
sum_x2y=x2_y-((x2_sum)*(y_sum))/df.shape[0]
sum_x1x2=x1_x2-((x1_sum)*(x2_sum))/df.shape[0]
```

## Calculating $B_1$ & $B_2$ 

```python
b1=((sum_x22)*(sum_x1y)-(sum_x1x2)*(sum_x2y))/((sum_x12)*(sum_x22)-(sum_x1x2)**2)
print("b1 = ",b1)
b2=((sum_x12)*(sum_x2y)-(sum_x1x2)*(sum_x1y))/((sum_x12)*(sum_x22)-(sum_x1x2)**2)
print("b2 = ",b2)
```

# Calculating $B_0$

```python
b0=y_mean-(b1*x1_mean)-(b2*x2_mean)
print("b0 = ",b0)
```

### Predicting Values Using The Obtained Parameters

```python
print("***********PRICE PREDICTOR******************")
print("Enter Sqft_Living")
sqft_living=float(input())
print("Enter Sqft_Lot")
sqft_lot=float(input())
price = b0 + (b1*sqft_living) + (b2*sqft_lot)
print("The Predicted Price Is: ", price )
```
