```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Simple Trignometric Functions

```python
def cosecant_plot():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    cosecantwave = 1/np.sin(x)
    plt.plot(x, cosecantwave)
    plt.ylim(-3, 3)
    plt.show()
    plt.plot(x, cosecantwave, "*",color="red")
    plt.ylim(-3, 3)
    plt.show()
```

```python
def cotangent_plot():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    cotangentwave = 1/np.tan(x)
    plt.plot(x, cotangentwave)
    plt.ylim(-3, 3)
    plt.show()
    plt.plot(x, cotangentwave, "*",color="red")
    plt.ylim(-3, 3)
    plt.show()
```

# Mixed Plots

```python
def mixed_cos_sec_plot():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    secwave = 1/np.cos(x)
    coswave = np.cos(x)
    plt.plot(x, coswave, x, coswave, "or", x, secwave, x, secwave, "og")
    plt.ylim(-3, 3)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Mixed Plot Of Cosecant And Secant Wave')
    plt.show()
    
```

# Mixed Plots With Legend

```python
    
def mix_plot_with_legend():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    tanwave = np.tan(x)
    cotwave = 1/np.tan(x)
    plt.plot(x, tanwave, label='tanwave', color ='red', linestyle = '--', linewidth =1.5)
    plt.ylim(-3, 3)
    plt.plot(x, cotwave, label='cotwave', color ='green', linestyle = '-', linewidth =1.5)
    plt.ylim(-3, 3)
    plt.legend()
    plt.show()
```

# Scatter Plot

```python
def scatterplot_fileread():
    travel_data = pd.read_csv('Travel_Location.csv')
    df = travel_data
    
    colors = {'Maldives':'g', 'Berlin':'r', 'Prague':'Blue'}
    plt.scatter(df['days'], df['expense'], c=df['location'].apply(lambda col_vector: colors[col_vector]), s = 200)
    plt.show()
```

```python
print("\n-----------------Welcome To Graph Plotting-------------------------------\n") 




while(True):
    print("1: Cosecant Plot \n2: Cotangent Plot \n3. Mix Cosine Secant Plot\n4. Mix Plot with Legend\n5. Scatter Plot and File Read\n6. Exit ")
    print("\n------------------------------------------------\n") 
    x=int(input("Enter your choice: "))
    if x==1:
        cosecant_plot()
    
    elif x==2:
        cotangent_plot()

    elif x==3:
        mixed_cos_sec_plot()

    elif x==4:
        mix_plot_with_legend()

    elif x==5:
        scatterplot_fileread()
        
    elif x==6:
        break

    else:
        print("Please enter valid choice")
```
